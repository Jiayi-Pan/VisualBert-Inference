```python
from vb_utils.visualbert_model import VB_Model
from PIL import Image
```

```python
vb_model = VB_Model(device='cuda')
```

```python
text = "A guy jumps in the air with his [MASK][MASK] beneath him"
img_dir = "coco_sample.jpg"
```

```python
Image.open(img_dir)
```

```python
out = vb_model.unmask(text, img_dir)
```

```python
predicted_cap =  vb_model.tokenizer.convert_ids_to_tokens(out['prediction_logits'].argmax(dim=-1)[0].tolist())[1:len(vb_model.tokenizer.tokenize(text))+1]
print(" ".join(predicted_cap))
```
